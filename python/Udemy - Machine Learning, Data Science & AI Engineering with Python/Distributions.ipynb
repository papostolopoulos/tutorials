# Examples of Data Distributions

## Uniform Distribution

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

values = np.random.uniform(-10.0, 10.0, 100000)
plt.hist(values, 50)
plt.show()
```

## Normal / Gaussian

Visualize the probability density function:

```python
from scipy.stats import norm
import matplotlib.pyplot as plt

# We're creating a list of x values for plotting that range between negative three and positive three, with a increment of 0.001 in between them.
x = np.arange(-3, 3, 0.001)
print(x)

# And then we're going to plot the x axis and the y axis is going to be the normal function norm.pdf  
# probability density function for a normal distribution on those x values.
plt.plot(x, norm.pdf(x))
plt.show()
```

Generate some random numbers with a normal distribution. "mu" is the desired mean, "sigma" is the standard deviation:

```python
import numpy as np
import matplotlib.pyplot as plt

mu = 5.0 #the mean
sigma = 2.0 #the standard deviation
values = np.random.normal(mu, sigma, 10000)
plt.hist(values, 50)
plt.show()
```

## Exponential PDF / "Power Law"

```python
from scipy.stats import expon
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.001)
plt.plot(x, expon.pdf(x))
plt.show()
```

## Binomial Probability Mass Function

```python
from scipy.stats import binom
import matplotlib.pyplot as plt

# This one deals with discrete data. The n, p values are the ones thad define how to shape the data.
n, p = 10, 0.5
x = np.arange(0, 10, 0.001)
print(n)
print(p)
print(x)
plt.plot(x, binom.pmf(x, n, p))
plt.show()
```
